```python
#01 WAP to add (append) value in this list
l1 = [10,20,30,40,50,60,70,80,90]
l1.append(100)
print(l1)
```

```python
#02 WAP to add list in this list
l1.append([1000,1500,2000])
print(l1)
```

```python
#03 WAP to insert a value in 3rd positon 
l1.insert(3,500)
print(l1)
```

```python
#04 WAP to insert a list in 5th position
l1.insert(5,[600,700,900])
print(l1)
```

```python
#05 WAP to extend a list 
l1.extend([2500,3500,4500])
print(l1)
```

```python
#06 WAP to remove last element 
l1.pop()
print(l1)
```

```python
#07 WAP to remove 5th element 
l1.pop(5)
print(l1)
```

```python
#08 WAP to remove 500
l1.remove(500)
print(l1)
```

```python
#09 WAP to print length of this list 
print(l1)
len(l1)
```

```python
#10 WAP to clear all element in this list 
l1.clear()
print(l1)
```
