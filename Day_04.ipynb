```python
#01 WAP to calculate factorial number using while loop
num=(int(input("Enter the Number")))
res=1
while(num>0):
    res=res*num
    num=num-1
print(res)
```

```python
#02 WAP to print 1 to 10 number using for loop
for i in range(1,11):
    print(i)
```

```python
#03 WAP to print 10 to 1 number using for loop
for i in range(10,0,-1):
    print(i)
```

```python
#04 WAP to print a number in only 6 steps
for i in range(30,100,6):
    print(i)
```

```python
#05 WAP to calculate multiplication table using for loop
num=(int(input("Enter the Number")))
for i in range(1,11):
    print(num*i)
```

```python
#06 WAP to calculate a 11 to 20 tables using for loop
for i in range(11,21):
    for num in range(1,11):
        print(i*num,end=" ")
    print()
```

```python
#07 WAP to print vowels in words 
word=(input("Enter the Word"))
for vowels in word:
    if vowels=='a' or vowels=='e' or vowels=='i' or vowels=='o' or vowels=='u':
        print(vowels)
```

```python
#08 WAP to print string using for loop
for i in "PUNE":
    print(i)
```

```python
#09 WAP to print capital word 
word=(input("Enter the Word"))
for i in word:
    print(i.upper(),end="")
```

```python
#10 WAP to print small word
word=(input("Enter the Word"))
for i in word:
    print(i.lower(),end="")
```
