```python
#03 WAP to find the largest element in a list.
list2 = [10,20,30,40,50,60,70,80,90]

largest = list2[0]

for i in list2:
    if i > largest:
        largest = i

print("Largest Element =", largest)
```

```python
#04 WAP to find the smallest element in a list.
list2 = [10,20,30,40,50,60,70,80,90]

smallest = list2[0]

for i in list2:
    if i < smallest:
        smallest = i

print("Smallest Element =", smallest)
```

```python
#06 WAP to calculate the sum of all elements in a list.
list2 = [10,20,30,40,50,60,70,80,90]

sum = 0

for i in list2:
    sum = sum+i

print("SUM =", sum)
```

```python
#07 WAP to calculate the subtraction of all elements in a list
list2 = [10,20,30,40,50,60,70,80,90]

sub = list2[0]

for i in list2[1:]:
    sub = sub - i

print("SUBTRACTION =", sub)
```

```python
#08 WAP to calculate the multiplication of all elements in a list
list2 = [10,20,30,40,50,60,70,80,90]

mul = 1

for i in list2:
    mul = mul * i

print("MULTIPLICATION =", mul)
```

```python
#09 WAP to calculate the average of all elements in a list
list2 = [10,20,30,40,50,60,70,80,90]

total = 0

for i in list2:
    total = total + i

average = total / len(list2)

print("AVERAGE =", average)
```

```python
#10 WAP to generate list of random numbers and make to different list as odd number list and even number list from original list 
list3 = [1,2,3,4,5,6,7,8,9]
odd=[]
even=[]
for i in list3:
    if i%2==0:
        even.append(i)
    else:
        odd.append(i)
print(f"Original list : {list3}\nEven number list: {even}\nOdd number list: {odd}")
```
