```python
#01 WAP to print right triangle pattern program
rows=(int(input("Enter the Number")))
for i in range(1,rows+1):
    print("*"*i)
```

```python
#02 WAP to print pattern program
rows=(int(input("Enter the Number")))
for i in range(1,rows+1):
    if(i%2==0):
        print("$"*i)
    else:
        print("#"*i)
```

```python
#03 WAP to print square pattern program
rows=(int(input("Enter the Number")))
for i in range(rows):
    for j in range(rows):
        print("*",end=" ")
    print()
        
```

```python
#04 WAP to print inverted right triangle pattern program
rows=(int(input("Enter the Number")))
for i in range(rows,0,-1):
    for j in range(i):
        print("*",end=" ")
    print()
```

```python
#05 WAP to print number triangle pattern program
rows=(int(input("Enter the Number")))
for i in range(1,rows+1):
    for j in range(1,i+1):
        print(j,end=" ")
    print()
```

```python
#06 WAP to print same number triangle pattern program
rows=(int(input("Enter the Number")))
for i in range(1,rows+1):
    for j in range(i):
        print(i,end=" ")
    print()
```

```python
#07 WAP to print continuous number triangle pattern program
rows=(int(input("Enter the Number")))
num=1
for i in range(1,rows+1):
    for j in range(i):
        print(num,end=" ")
        num=num+1
    print()
```

```python
#08 WAP to print reverse number triangle pattern program
rows=(int(input("Enter the Number")))
for i in range(rows,0,-1):
    for j in range(i):
        print(i,end=" ")
    print()
```

```python
#09 WAP to print reverse continuous number triangle pattern program
rows=(int(input("Enter the Number")))
num=1
for i in range(rows,0,-1):
    for j in range(1,i+1):
        print(j,end=" ")
        num=num+1
    print()
```

```python
#10 WAP to print reverse continuous number triangle pattern program
rows=(int(input("Enter the Number")))
num= rows * (rows+1) // 2 
for i in range(rows,0,-1):
    for j in range(i):
        print(num,end=" ")
        num=num-1
    print()
```
