```python
#01 WAP to print Palindrome Number
n = int(input("Enter a number: "))

original = n
reverse = 0

while n > 0:
    digit = n % 10
    reverse = reverse * 10 + digit
    n = n // 10

if original == reverse:
    print("Palindrome Number")
else:
    print("Not a Palindrome Number")
```

```python
#02 WAP to print Armstrong Number
n = int(input("Enter a number: "))
temp = n
sum = 0

while temp > 0:
    digit = temp % 10
    sum = sum + digit ** 3
    temp = temp // 10

if sum == n:
    print("Armstrong Number")
else:
    print("Not Armstrong Number")
```

```python
#03 WAP to print Reverse Number
n = int(input("Enter a number: "))

reverse = 0

while n > 0:
    digit = n % 10
    reverse = reverse * 10 + digit
    n = n // 10

print("Reverse Number =", reverse)
```

```python
#04 WAP to print Reverse Number
n = int(input("Enter a number: "))
count = 0

while n > 0:
    count += 1
    n = n // 10

print(count)
```

```python
#05 WAP to print product of digits
n = int(input("Enter a number: "))
product = 1

while n > 0:
    digit = n % 10
    product = product * digit
    n = n // 10

print(product)
```

```python
#06 WAP to print Strong Number
n = int(input("Enter a number: "))

temp = n
sum = 0

while temp > 0:
    digit = temp % 10

    fact = 1
    i = 1

    while i <= digit:
        fact = fact * i
        i = i + 1

    sum = sum + fact
    temp = temp // 10

if sum == n:
    print("Strong Number")
else:
    print("Not Strong Number")
```

```python
#07 WAP to print Fibonacci Series
n = int(input("Enter number of terms: "))

a = 0
b = 1
i = 1

while i <= n:
    print(a, end=" ")

    c = a + b
    a = b
    b = c

    i = i + 1
```

```python
#08 WAP to print Spy Number
n = int(input("Enter Number: "))
temp = n
sum = 0
mul = 1

while temp > 0:
    digit = temp % 10
    sum = sum + digit
    mul = mul * digit
    temp = temp // 10

if sum == mul:
    print("Spy Number")
else:
    print("Not Spy Number")
```

```python
#09 WAP to print LCM of two number
a=int(input("Enter the Number"))
b=int(input("Enter the Number"))
max_num = max(a,b)
while True:
    if max_num % a == 0 and max_num % b == 0 :
        print("LCM =" ,max_num)
        break
    max_num += 1
```
