```python
#01 WAP to print upper Daimond pattern program
rows = int(input("Enter the Number"))

# Upper Part
for i in range(1, rows + 1):

    for j in range(rows - i):
        print(" ", end=" ")

    for j in range(2 * i - 1):
        print("*", end=" ")

    print()
    
```

```python
#02 WAP to print lower pattern program
rows = int(input("Enter the Number"))

# Lower Part
for i in range(rows-1,0,-1):

    for j in range(rows - i):
        print(" ", end=" ")

    for j in range(2 * i - 1):
        print("*", end=" ")

    print()
```

```python
#03 WAP to print Daimond pattern program
rows = int(input("Enter the Number"))

# Upper Part
for i in range(1, rows + 1):

    for j in range(rows - i):
        print(" ", end=" ")

    for j in range(2 * i - 1):
        print("*", end=" ")

    print()

# Lower Part
for i in range(rows-1,0,-1):

    for j in range(rows - i):
        print(" ", end=" ")

    for j in range(2 * i - 1):
        print("*", end=" ")

    print()
```

```python
#04 WAP to print square pattern using for loop
rows=int(input("Enter the Number"))
for i in range(rows):
    for j in range(rows):
        if i == 0 or i == rows-1 or j == 0 or j == rows-1:
            print("*",end=" ")
        else:
            print(" ",end=" ")
    print()
```

```python
#05 WAP to print traingle pattern using for loop
rows=int(input("Enter the Number"))
for i in range(rows+1):
    for j in range(1,i+1):
        if j == 1 or j == i or i == rows:
            print("*",end=" ")
        else:
            print(" ",end=" ")
    print()
```

```python
#06 WAP to print upper part of butterfly
n = int(input("Enter the Number"))

for i in range(1, n + 1):
    for j in range(i):
        print("*", end="")

    a = 2 * (n - i)

    for j in range(a):
        print(" ", end="")

    for j in range(i):
        print("*", end="")

    print()
```

```python
#07 WAP to print lower part of butterfly
n = int(input("Enter the Number"))

for i in range(n,0,-1):
    for j in range(i):
        print("*", end="")

    a = 2 * (n - i)

    for j in range(a):
        print(" ", end="")

    for j in range(i):
        print("*", end="")

    print()
```
